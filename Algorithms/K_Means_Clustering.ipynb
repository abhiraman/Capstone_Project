```python
import sklearn
from sklearn.datasets  import make_blobs
import seaborn as sns
```

```python
X, y_true = make_blobs(
    n_samples=400, 
    centers=4,
    cluster_std=0.60, 
    random_state=0
)
```

```python
import matplotlib
ax = matplotlib.pyplot.gca()
#m = X.reshape((X.shape[1],X.shape[0]))
ax = sns.scatterplot(X[:, 0], X[:, 1],palette = sns.color_palette("pastel",n_colors=4),ax = ax)
ax.set_xlabel("X1")
ax.set_ylabel("X2")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
n_clusters = 4
```

```python
estim = KMeans(n_clusters= n_clusters, random_state=0).fit(X)
```

```python
y_kmeans  = estim.predict(X)
```

```python
## Plot of Training data after fitting the input
sns.scatterplot(X[:, 0], X[:, 1],hue=y_kmeans,palette = sns.color_palette("pastel",n_colors=n_clusters))
```

```python
centriods = estim.cluster_centers_
print("centriods :",centriods)
sns.scatterplot(X[:, 0], X[:, 1],hue=y_kmeans,palette = sns.color_palette("pastel",n_colors=n_clusters))

plt.scatter(
    centriods[:, 0], 
    centriods[:, 1], 
    c='black',
    s=200,
    alpha=0.55
)
```

```python
sse = []
for k in range(1, 8):
    km = KMeans(n_clusters=k)
    km.fit(X)
    sse.append(km.inertia_)

# plot to identify the elbow
plt.figure(figsize=(6, 6))
plt.plot(range(1, 8), sse, '-o', c = 'maroon')
plt.xlabel('Count of Clusters')
plt.ylabel('SSE');
```
